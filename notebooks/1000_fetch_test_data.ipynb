```python
import os
# change working directory to ../src
os.chdir('../src')
print(f"Working directory: {os.getcwd()}")

import ast
from dotenv import load_dotenv
load_dotenv()

import pandas as pd

from utils.database import *

fs, db = getConnection(use_dotenv=True)
QUERY = ast.literal_eval(os.getenv("QUERY"))
FIELDS = ast.literal_eval(os.getenv("FIELDS"))
LIMIT = ast.literal_eval(os.getenv("LIMIT"))
SAMPLING = ast.literal_eval(os.getenv("SAMPLING"))
```

## fetchTasks

```python
tasks = fetchTasks(db, QUERY, FIELDS, LIMIT, SAMPLING)
tasks
```

```python
df_tasks = pd.DataFrame(tasks)
df_tasks
```

```python
# save to csv
os.makedirs('../data', exist_ok=True)
df_tasks.to_csv('../data/tasks.csv', index=False)
```

## fetchAnnotations

```python
annotations = db.pages.annotations.find()
df_annotations = pd.DataFrame(annotations)
df_annotations
```

```python
# save to data/annotations.csv
df_annotations.to_csv('../data/annotations.csv', index=False)
```

## fetchHTML

```python
# create html folder
os.makedirs('../data/html', exist_ok=True)
```

```python
# for each task, get the html content
for task in tasks:
    if not task['content_requests']:
        continue

    try:
        html = getPageContent(fs, task['content_requests'])
        with open(f"../data/html/{task['_id']}.html", 'w') as f:
            f.write(html)
    except Exception as e:
        print(f"Error: {e} for task {task['_id']}")
        continue
```
